```python
def prev_trdae (st1, instu_id,lot_size) :

  response = xt.get_position_netwise(
                                    clientID=clientID
                                    )

  trade_detail = 0
  script = st1
  scrip_id = instu_id
  quantity = lot_size
  buy_price = 0
  sell_price = 0
  #print(response)
  posit = response['result']['positionList']
  #print(1)
  if ( len(posit) > 0 ) :
    for pos in range(len(posit)) :
      if ( int(posit[pos]['ExchangeInstrumentId']) == scrip_id ) :

        if ( int(posit[pos]['Quantity']) == 0  ) :

          trade_detail = 0

        if ( int(posit[pos]['Quantity']) ==  quantity)   :
          trade_detail = 1
          buy_price = posit[pos]['BuyAveragePrice']


        if ( int(posit[pos]['Quantity']) == -quantity)   :

          trade_detail = -1
          sell_price = posit[pos]['SellAveragePrice']




  if ( len(posit) == 0 ) :
    trade_details = 0


  return trade_detail, buy_price , sell_price
```

```python
def order_aftermath (order_id) :

  #order history
  response = xt.get_order_history(
                        appOrderID = order_id,
                        clientID = clientID
                                       )

  # order execution
  order_stat = 0
  order_status = []
  reason = []
  leave_qty = 0
  order_list = response['result']
  if len(order_list) > 0:
    for i in range ( len(order_list) ) :
      if order_list[i]['AppOrderID'] == order_id :
        order_stat == 1
        order_status.append(order_list[i]['OrderStatus'])
        reason.append(order_list[i]['CancelRejectReason'])
        leave_qty = order_list[i]['LeavesQuantity']

      else  :
        order_stat == 0
        order_status.append(order_list[i]['OrderStatus'])

  if len(order_list) < 0 :
    order_status.append('NO ORDERS')

  return order_status , reason , leave_qty
```
